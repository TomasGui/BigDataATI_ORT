```python
#importacion de paquetes 
import pandas as pd 
import matplotlib.pyplot as plt
import pydeck as pdk
import os
import streamlit as st
from datetime import datetime
```

```python
#Data Sets
#Data Set Fast test Laps 
df_fastest_laps = pd.read_csv("output_data/fastest_laps_clean.csv") 
#Data Set Teams
df_teams = pd.read_csv("output_data/teams_clean.csv")
#Data Set Drivers
df_drivers = pd.read_csv("output_data/drivers_clean.csv")
#Data Set Winners - Revisar
df_winners = pd.read_csv("output_data/winners_clean.csv")
```

<h2>           ¿Cuáles son los mejores tiempos promedio por circuito?        </h2>

```python
#Datos promedios por circuitos - Circuito es Grand Prix 
#Ajuste de tipo de  Datos TIME esta String
# Función para convertir el tiempo a segundos
def time_to_seconds(time_str):
    if ':' in time_str:
        minutes, seconds = time_str.split(':')
        return int(minutes) * 60 + float(seconds)
    else:
        return None  # or some other handling for incorrect format

# Aplicar la función a la columna 'Time'
df_fastest_laps['Time_sec'] = df_fastest_laps['Time'].apply(time_to_seconds)

# Filtrar filas con tiempos válidos
df_fastest_laps = df_fastest_laps[df_fastest_laps['Time_sec'].notna()]

# Calcular el tiempo promedio por circuito
mean_times = df_fastest_laps.groupby('Grand Prix')['Time_sec'].mean().reset_index()

# Convertir los tiempos promedio de vuelta a formato minutos:segundos
def seconds_to_time(seconds):
    minutes = int(seconds // 60)
    sec = seconds % 60
    return f"{minutes}:{sec:.3f}"

mean_time_timeSec = mean_times
mean_time_timeSec['Avg Time'] = mean_time_timeSec['Time_sec'].apply(seconds_to_time)

# Mostrar los resultados
#print(mean_time_timeSec[['Grand Prix', 'Avg Time']])
```

```python
# Graficar el resultado 
plt.figure(figsize=(12, 8))
plt.scatter(mean_times["Grand Prix"], mean_times["Time_sec"], color='green')
plt.title('Tiempo promedio por circuito', fontsize=16)
plt.xlabel('Grand Prix - Circuito', fontsize=14)
plt.ylabel('Avg Time - Promedio (Segundos)', fontsize=14)
plt.xticks(rotation=90, fontsize=9)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<h2>           ¿Cuáles son los autos con los mejores tiempos?       </h2>

```python
# Calcular el tiempo promedio por Auto
mean_times_car = df_fastest_laps.groupby('Car')['Time_sec'].mean()
mean_times_car = mean_times_car.sort_values(ascending = True)
#Solo vamos a tomar los 10 mjores autos
lista_autos = mean_times_car[:10]
# Graficar el resultado 
plt.figure(figsize=(12, 8))
plt.scatter(lista_autos.index, lista_autos.values, color='red')
plt.title('Tiempo promedio por autos', fontsize=16)
plt.xlabel('Autos', fontsize=14)
plt.ylabel('Avg Time - Promedio (Segundos)', fontsize=14)
plt.xticks(rotation=90, fontsize=9)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<h2>¿Quiénes son los pilotos con mayor desempeño en los circuitos, basado en sus tiempos?</h2>

```python
# Calcular el tiempo promedio por Auto
#Mejor tiempo se entiende por el minimo valor que tiene un piloto
mean_times_driver = df_fastest_laps.groupby('Driver')['Time_sec'].min()
mean_times_driver = mean_times_driver.sort_values(ascending = True)
#Solo vamos a tomar los 10 mejores tiempos
lista_pilotos = mean_times_driver[:10]
# Graficar el resultado 
plt.figure(figsize=(12, 8))
plt.scatter(lista_pilotos.index, lista_pilotos.values, color='red')
plt.title('Mejores tiempos por Piloto', fontsize=16)
plt.xlabel('Pilotos', fontsize=14)
plt.ylabel('Tiempo (Segundos)', fontsize=14)
plt.xticks(rotation=90, fontsize=9)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<h2>¿Cómo se pueden listar los circuitos y los pilotos en un mapa?</h2>


```python
#Crear un nuevo dataframe que contenga los datos del circuito y su ubicacion 
# Seleccionar columnas específicas
columnas_seleccionadas = ['Grand Prix', 'Latitud', 'Longitud']
df_seleccionado = df_fastest_laps[columnas_seleccionadas]
# Eliminar duplicados basados en una columna específica
df_seleccionado = df_seleccionado.drop_duplicates(subset='Grand Prix')
df_seleccionado = df_seleccionado.dropna()
#Confeccion del mapa 
layer = pdk.Layer(
    'ScatterplotLayer',  # `type` positional argument is here
    df_seleccionado,
    get_position=['Longitud', 'Latitud'],
    auto_highlight=True,
    get_radius=80000,          # Radius is given in meters
    get_fill_color=[180, 0, 200, 140],  # Set an RGBA value for fill
    pickable=True)

# Set the viewport location
view_state = pdk.ViewState(
    longitude=0,    # Longitud central
    latitude=0,     # Latitud central
    zoom=0.5,         # Nivel de zoom bajo para vista global
    min_zoom=0,
    max_zoom=15,
    pitch=0,
    bearing=0)

#Estilo de mapa por medio de un combo box
mapstyle = st.sidebar.selectbox(
    "Selecciona estilo:",
    options=["light", "dark", "satellite", "road"],
    format_func=str.capitalize,
)

# Combined all of it and render a viewport
r = pdk.Deck(layers=[layer], initial_view_state=view_state , map_style="road")
r.to_html('Circuitos_mapa.html')
```

<h2>           ¿Cuáles son los autos con más carreras ganadas?       </h2>

```python
#Obtener autos ordenados por cantidad de victorias
winner_cars = df_winners['Car'].value_counts()
#Solo vamos a tomar los 10 mjores autos
lista_autos_ganadores = winner_cars[:10]
# Graficar el resultado
plt.figure(figsize=(12, 8))
plt.scatter(lista_autos_ganadores.index, lista_autos_ganadores.values, color='red')
plt.title('Autos con más carreras ganadas', fontsize=16)
plt.xlabel('Autos', fontsize=14)
plt.ylabel('Victorias', fontsize=14)
plt.xticks(rotation=90, fontsize=9)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<h2>           ¿Cuál es el promedio de mejores tiempos por año?       </h2>

```python
#Genero dataframe especifico para esta consulta
df_mean_times_year = df_winners.copy()
#Agrego columna con año al dataframe
df_mean_times_year['Date'] = pd.to_datetime(df_mean_times_year['Date'])
df_mean_times_year['Year'] = df_mean_times_year['Date'].dt.year
#Convierto campo Time a segundos
df_mean_times_year['Time_sec'] = df_mean_times_year['Time'].apply(time_to_seconds)
# Filtrar filas con tiempos válidos
df_mean_times_year = df_mean_times_year[df_mean_times_year['Time_sec'].notna()]
# Calcular el tiempo promedio por año
mean_times_year = df_mean_times_year.groupby('Year')['Time_sec'].mean().reset_index()
# Convertir los tiempos promedio de vuelta a formato horas:minutos:segundos
def seconds_to_time_with_hours(seconds):
    hours = int(seconds // 3600)
    minutes = int(seconds % 3600) // 60
    sec = seconds % 60
    return f"{hours}:{minutes}:{sec:.3f}"

mean_times_year

#Agrego columna Avg Time con tiempos promdedios del año
mean_times_year['Avg Time'] = mean_times_year['Time_sec'].apply(seconds_to_time_with_hours)
mean_times_year = mean_times_year.sort_values(by=['Time_sec'])

#Solo vamos a tomar los 10 mjores autos
best_years = mean_times_year[:10]
best_years

# Graficar el resultado 
plt.figure(figsize=(12,8))
plt.scatter(best_years["Year"], best_years["Avg Time"], color='green')
plt.title('Mejores tiempos promedio por año', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Tiempo Promedio', fontsize=14)
plt.xticks(rotation=90, fontsize=9)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<h2>           ¿Cuáles son los teams con más premios en los últimos 5 años? ¿Y en los últimos 10?       </h2>


```python
#Genero dataframe especifico para esta consulta
df_best_teams_years = df_winners.copy()
#Agrego columna con año al dataframe
df_best_teams_years['Date'] = pd.to_datetime(df_mean_times_year['Date'])
df_best_teams_years['Year'] = df_mean_times_year['Date'].dt.year
#Filtro 5 años de datos
df_best_teams_5_years = df_best_teams_years[df_best_teams_years['Year'].between(datetime.now().year-5, datetime.now().year)]
#Filtro 10 años de datos
df_best_teams_10_years = df_best_teams_years[df_best_teams_years['Year'].between(datetime.now().year-10, datetime.now().year)]
#Obtener autos ordenados por cantidad de victorias
winner_cars_10_years = df_best_teams_10_years['Car'].value_counts()
#Solo vamos a tomar los 10 mjores autos
lista_autos_ganadores_10_años = winner_cars_10_years[:10]
# Graficar el resultado
plt.figure(figsize=(12, 8))
plt.scatter(lista_autos_ganadores_10_años.index, lista_autos_ganadores_10_años.values, color='red')
plt.title('Autos con más carreras ganadas en últimos 10 años', fontsize=16)
plt.xlabel('Autos', fontsize=14)
plt.ylabel('Victorias', fontsize=14)
plt.xticks(rotation=90, fontsize=9)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
